# Matplotlib: plotting 03

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Histograms

Building histograms:

```python
mean = 500
std = 50
size = 1000
data = np.random.normal(loc=mean, scale=std, size=size)
```

```python
plt.hist(data)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("Freq")
```

Personalizing color and number of bins:

```python
n_bins = 20
plt.hist(data, n_bins, facecolor='green', alpha=0.5)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("Freq")
```

```python
n_bins = 20
plt.hist(data, n_bins, normed=1, facecolor='green', alpha=0.5)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("density")
```

Adding a line showing the expected distribution:

```python
import matplotlib.mlab as mlab

n_bins = 20
n, bins, patches = plt.hist(data, n_bins, normed=1, facecolor='green', alpha=0.5)
print(bins)
y = mlab.normpdf(bins, mean, std)
l = plt.plot(bins, y, 'r--', linewidth=1.5)
plt.title("Histogram")
plt.xlabel("Data")
plt.ylabel("density")
plt.subplots_adjust(left=0.15)
```
