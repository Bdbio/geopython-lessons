# Vector analysis with GDAL (OGR)

Vector analisys examples with OGR:

```python
from osgeo import ogr
```

```python
geol_fl = "../../data/geologia_marina/mm01_geologia.shp"
```

Reading a dataset:

```python
ds = ogr.Open(geol_fl, 0)
if ds is None:
    sys.exit('Error opening file')
```

```python
lyr = ds.GetLayer(0)
```

```python
feat = lyr.GetFeature(0)
```

```python
geom = feat.geometry()
```

```python
geom.ExportToWkt()
```

Computing geometry centroid:

```python
geom.Centroid().ExportToWkt()
```

Computing geometry area:

```python
geom.Area()
```

Computing buffer geometry at a given distance:

```python
dist = 10000 # meters
buff = geom.Buffer(dist)
```

```python
buff.ExportToWkt()
```
